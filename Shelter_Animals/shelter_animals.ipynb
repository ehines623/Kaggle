# Features

Let's make a new data frame with only the features to be fed into the ML

```python
train_final = pd.DataFrame()
```

```python
train_final['HasName'] = train['Name'].notnull()
```

```python
train_final['IsPitbull'] = train['Breed'].str.contains('pit', case=False)
```

```python
train_final['IsMix'] = (train['Breed'].str.contains('mix', case=False) | train['Breed'].str.contains('/'))
```

```python
animal_types = list(enumerate(np.unique(train['AnimalType'])))
animal_dict = { name: i for i, name in animal_types }
train_final['AnimalType'] = train['AnimalType'].map(lambda x: animal_dict[x]).astype(int)
```

```python
#sex_types = list(enumerate(np.unique(train['SexuponOutcome'])))  #defined above for plotting
sex_dict = { name: i for i, name in sex_types }
train_final['SexType'] = train['SexuponOutcome'].map(lambda x: sex_dict[x]).astype(int)
```

Make sure the date is in datetime, so we can do more useful things with it

```python
train['DateTime'] = pd.to_datetime(train['DateTime'], format='%Y-%m-%d %H:%M:%S')
train_final['year'], train_final['month'] = train['DateTime'].apply(lambda x: x.year), train['DateTime'].apply(lambda x: x.month)

train_final.head()
```

```python
train['AgeuponOutcome'].value_counts()
```

```python
def parseAge(age):
    #print type(age)
    if age=='Unknown':
        return 'Unknown'
    (num, time_type) = str.split(age)
    num = int(num)
    if 'day' in time_type:
        if num < 56:
            return 'Infant'
        elif num <365:
            return 'Baby'
        else:
            return 'Error'
        
    elif 'week' in time_type:
        if num < 8:
            return 'Infant'
        elif num < 52:
            return 'Baby'
        else:
            return 'Error'
        
    elif 'month' in time_type:
        if num < 2:
            return 'Infant'
        elif num < 12:
            return 'Baby'
        elif num < 36:
            return 'Young'
        else:
            return 'Error'
        
    elif 'year' in time_type:
        if num <1:
            return 'Baby'
        elif num<3:
            return 'Young'
        elif num<8:
            return 'Adult'
        else:
            return 'Senior'   
    else:
        return 'Error'
    

```

These categories based on those used on Petfinder.com (adding in the infant)

Infant -- anything less than 8 weeks (ie should not be leaving mother) 

Baby -- less than 1 year

Young -- less than 3 years

Adult -- 3 - 8 years

Senior 8+ years (this is probably slightly too low for cats, but could be tuned later)

```python
train_final['AgeCategory'] = train['AgeuponOutcome'].map(parseAge)
```

```python
#age_types = list(enumerate(np.unique(train['AgeCategory'])))
age_types = [(0,'Infant'),(1,'Baby'),(2,'Young'),(3,'Adult'),(4,'Senior'),(5,'Unknown')]
age_dict = { name: i for i, name in age_types }
train_final['AgeCategory'] = train_final['AgeCategory'].map(lambda x: age_dict[x]).astype(int)
```

May be worth investigating this a bit more to see if defining categories works better or allowing it to be a continuous variable is better

# Do some wrangling to get everything in the format needed for input

```python
train_final.head()
```

```python
train_outcomes = pd.DataFrame()
```

```python
train_outcomes['Adoption'] = (train['OutcomeType'] == 'Adoption').astype(int)
train_outcomes['Died'] = (train['OutcomeType'] == 'Died').astype(int)
train_outcomes['Euthanasia'] = (train['OutcomeType'] == 'Euthanasia').astype(int)
train_outcomes['Return_to_owner'] = (train['OutcomeType'] == 'Return_to_owner').astype(int)
train_outcomes['Transfer'] = (train['OutcomeType'] == 'Transfer').astype(int)

outcomes_columns = train_outcomes.columns

train_outcomes = train_outcomes.as_matrix()
```

```python
outcomes_columns
```

# Train the Multiclass Random Forest

```python
forest = OneVsRestClassifier(RandomForestClassifier(n_estimators = 100))
```

```python
forest = forest.fit(train_final,train_outcomes)
```

```python
for classifier in range(len(forest.estimators_)):
    importances = forest.estimators_[classifier].feature_importances_
    std = np.std([tree.feature_importances_ for tree in forest.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    # Print the feature ranking
    print("Feature ranking for %s estimator:" % outcomes_columns[classifier])

    for f in range(train_final.shape[1]):
        print("%d. %s (%f)" % (f + 1, train_final.columns[indices[f] ], importances[indices[f]]))
    print
```
